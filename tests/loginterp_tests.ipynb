```python
import numpy as np

from matplotlib import pyplot as plt
```

```python
# Make trial function
x = np.logspace(-2,2,10**2)
y = - (x - 1)**2 + 10
```

```python
plt.loglog(x, np.abs(y), '.')
```

```python
plt.semilogx(x,np.abs(y))
plt.ylim(-1,11)
```

```python
# Find the zero:
ii = np.argmin(np.abs(y))
print(y[ii-1], y[ii], y[ii+1])
```
