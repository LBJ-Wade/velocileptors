```python
import numpy as np
from loginterp import loginterp

import sys

from matplotlib import pyplot as plt
```

```python
pks = np.loadtxt('../zvonimir/ds14_a_1.0000_pk.dat')
k = pks[:,0]; p = pks[:,1]
```

```python
import time
```

```python
from velocity_moments_expanded_fftw import EVelocityMoments
```

```python
start = time.time()
ezelda = EVelocityMoments(k, 0.5*p, beyond_gauss=True, one_loop= True, third_order = True, import_wisdom=False, extrap_min=-5,extrap_max=3,N=2000)
end = time.time()
print(end-start)
```

```python
start = time.time()
ezelda.make_ptable()
end = time.time()
print(end-start)
```

```python
pktable_old = np.loadtxt('./cleft_pktable.dat')
```

```python
for ii in range(12):
    plt.title(str(ii))
    plt.loglog(ezelda.pktable[:,0],np.abs(ezelda.pktable[:,ii+1]) )
    plt.loglog(pktable_old[:,0], np.abs(pktable_old[:,ii+1]), 'k--')
    plt.show()
```

```python
ii = 0
plt.plot(ezelda.pktable[:,0], ezelda.pktable[:,0] * np.abs(ezelda.pktable[:,ii+1]) )
plt.plot(pktable_old[:,0], pktable_old[:,0]*np.abs(pktable_old[:,ii+1]), 'k--')
plt.xlim(0,0.5)
plt.ylim(0,4e2)
plt.show()
```

```python
plt.plot(ezelda.pktable[:,0], ezelda.pktable[:,ii+1]/pktable_old[:,ii+1] - 1)
plt.plot(pktable_old[:,0], -0.1*pktable_old[:,0]**4)
plt.xscale('log')
plt.xlim(1e-3,1)
plt.ylim(-0.01,0.01)
plt.show()
```
